```python
from warnings import filterwarnings
filterwarnings('ignore')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster.elbow import kelbow_visualizer
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import plotly.express as px

```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

### Categorize similar customers on basis of Age , Income and spending Score

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.columns = ['CustomerID', 'Gender', 'Age','Income','Spend']
```

```python
df.head()
```

### Seperate x data

```python
x = df[['Age','Income','Spend']]
x.head()
```

### Process the x data

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler',StandardScaler())]).set_output(transform='pandas')
```

```python
x_pre = num_pipe.fit_transform(x)
x_pre.head()
```

```python
x_pre.mean()
```

```python
x_pre.std()
```

### Calculate optimal number of clusters on basis of Silhoutee score

```python
n = list(range(2,15))
n
```

```python
scores = []
for i in n:
    model = AgglomerativeClustering(n_clusters=i)
    model.fit(x_pre)
    # Predict the model clusters
    lb = model.labels_
    # calculate silhoutee scores
    sil = silhouette_score(x_pre,lb)
    scores.append(sil)
```

```python
scores
```

### Visualize silhoutee scores

```python
plt.figure(figsize=(12,6))
plt.plot(n,scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhoutee Score')
plt.title('Silhoutee Scores for various clusters')
plt.grid()
plt.show()
```

### Based on above graph for 6 clusters we have highest score of 0.42
Create a Agglomerative model with n_clusters=6

```python
model = AgglomerativeClustering(n_clusters=6)
model.fit(x_pre)
```

```python
model.labels_
```

### Save the above results in x

```python
x['Label'] = model.labels_
x.head()
```
